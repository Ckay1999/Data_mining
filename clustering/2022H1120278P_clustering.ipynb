### Using elbow nick method

```python
sse = {}
for k in range(1,11 ):
    kmeans = KMeans(n_clusters = k, n_init = 10, max_iter = 300, random_state = 42)
    kmeans.fit(X)
    sse[k] = kmeans.inertia_
print(sse)
```

```python
%matplotlib inline
import matplotlib.style as style

x = [i for i in range(1, 11)]
y = list(sse.values())

style.use('fivethirtyeight')
plt.plot(x, y)
plt.xlabel("Value of k in KMeans")
plt.ylabel("SSE")
plt.xticks(x, list(sse.keys()))
plt.title("k versus SSE Plot")
plt.show()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as shc
from sklearn import metrics
%matplotlib inline
from sklearn.preprocessing import normalize
from sklearn.decomposition import PCA
```

```python
data=pd.read_csv("s1_modified.csv")
data=data.head(1000)
```

```python
data.head(5)
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(data)
```

```python
db_default = DBSCAN(eps = 0.3, min_samples = 50).fit(X_scaled)
labels = db_default.labels_
```

```python
plt.scatter(data.iloc[:,0].values,data.iloc[:,1].values,c=labels,cmap='plasma')
```

```python
n_clusters_=len(set(labels))-(1 if -1 in labels else 0)
n_noise_=list(labels).count(-1)
n_clusters_,n_noise_
```

```python
outliers=data[labels==-1]
print(outliers)
plt.scatter(outliers.x,outliers.y)
```
